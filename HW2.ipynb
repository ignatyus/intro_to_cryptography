**Question:** Let $p$ be an odd prime and let $g$ be a primitive root modulo $p$. Prove that $a$ has a square root modulo $p$ if and only if its discrete logarithm $log_g(a)$ modulo $p−1$ is even.

**Solution:**

Now $p$ be an odd prime and let $g$ be a primitive root modulo $p$.
We want to prove that $a$ has a square root modulo $p$ if and only if its discrete logarithm $log_g(a)$ modulo $p−1$ is even.

**Proof of only if part:**

Assume that $a$ has a square root module $p$. This means that there exist an integer $b$ such that $b^2 \equiv a \text{ (mod p) }$  

Then, let $k = log_g(b) \text{ (mod p)}$, for $ k \in [0, p-1]$. Then we have $ g^{2k} = b^2 = a$ Let $ 2k=n(p−1)+m$ for $ n \in \mathbb{Z}$ and $ m \in [0, p-1]$ Then we have $ g^m = g^{2k−n(p−1)} = \frac{g^{2k} }{g^{n(p-1)}} $ which is equal to $ g^{2k} $ since $g^{n(p-1)}$ is equal to 1 by FTL. This implies $ g^{2k} = a \text{ (mod p)}$, that is to say, $log_g(a) = m = 2k$, hence $log_g(a)$ is even.


**Proof of if part:**

Let p be an odd prime and let g be a primitive root modulo p. Let $log_g(a) \text{ (mod p−1) }$ be even. Then, $g^{2k} = a$, for some $k \in \mathbb{Z} $. This implies that there exist a b such that $b^2 = a$, specifically for $b = g^k$. Hence, a has a square root modulo p.

$$ x \equiv \pm a \text{ (mod p)} $$

We have that s is not 0 modulo n, we have this sum s is divisible by either p or by q such that $gcd(n,s) = p$ or $gcd(n,s) = q$, which are the factors of n. **Question:** Use the Miller–Rabin test to provide a Miller–Rabin witness for the compositeness of n, or conclude that n is probably prime by providing 10 numbers that are not Miller–Rabin witnesses for n.

$$ n = 118901527 $$

**Solution:**

For this question, we are going to write and use an algorithm that tests potential Miller-Rabin witnesses:


```python
# integer n to bested, integer a as potential witness
def MillerRabinTest(n, a):
    # 1
    if (n % 2 == 0 or 1 < gcd(a, n) < n):
        return True # composite
    # 2 write n-1 = 2^{k}q with q odd
    q = n - 1
    k = 0
    while(q % 2 == 0):
        q = q / 2
        k = k + 1
    
    # 3 . Set a = a^{q} mod n
    a = Mod(a, n)^q
    
    # 4. If a is 1 mod n then test fails
    if(Mod(a, n) == 1):
        return False
    
    # 5. 
    for i in range(k):
        # 6.
        if( Mod(a, n) == Mod(-1, n) ):
            return False
        # 7. Set a = a^2 mod n
        a = Mod(a, n)^2
    
    # 9
    return True # composite
```

So we need to test $n$ with numbers $a$ that are not divisible by $n$. Lets test it with 10 random integers between 1 and n so n cannot divide them.




```python
n = 118901527
mrandom_integers = [choice(range(1, 100)) for i in range(10)]
print("Our randomly choosen integers:", mrandom_integers)

foundWitness = False
for a in mrandom_integers:
    if( MillerRabinTest(n, a)):
        foundWitness = True
        print(a, " is a Miller-Rabin witness for the number ", n)
if( not foundWitness):
    print("None of the randomly choosen 10 numbers is a Miller-Rabin witness for", n)
```

Thus we can conclude that 118901527 is (probably) a prime number.

**Question:** Use Pollard’s p − 1 method to factor $ n = 48356747 $

**Solution:** 

By using the pollardsFactorization method that is described belove, we found our two factors as 6917 and 6991.


```python
# input: Integer N to be factored
def pollardsFactorization(N):
    # 1. Set a = 2
    a = 2
    # 2. Loop j = 2,3,4.. up to a specified bound
    j = 2
    while(True):
        # 3. set a = a^j mod N
        a = a^j % N
        # 4. compute gcd(a-1, N)
        d = gcd(a - 1, N)
        # if 1 < d < N, then success
        if(1 < d < N):
            return d
        
        j = j + 1
        
print("Our first factor is:  ", pollardsFactorization(48356747)) # our first factor
print("Our second factor is: ", 48356747/6917 ) # our second factor
```
